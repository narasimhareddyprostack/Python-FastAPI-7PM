```python
emp={}
emp={'eid':101,'ename':'Rahul','esal':45000}
print(emp)
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}
#how to read dict values - using dict[key]
print(emp['eid'])     #101
print(emp['ename'])   #Rahul
print(emp['esal'])    #450000
print(emp['loc'])     #KeyError
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}
#update
emp['ename']= "Rahul Gandhi"
#Delete
del emp['esal']
print(emp)
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}
print(emp.get('eid'))  #101
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}
print(emp.get('loc'))  #None
print(emp['loc'])      #KeyError
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}
values=emp.values()
print(values)

for value in values:
    print(value)
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}


for value in emp.values():
    print(value)
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}

for key in emp.keys():
    print(key)
```

```python
emp={'eid':101,'ename':'Rahul','esal':45000}

for key,value in emp.items():
    print(key,":",value)
```
